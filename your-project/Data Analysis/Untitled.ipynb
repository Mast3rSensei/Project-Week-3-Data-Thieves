```python
import pandas as pd
import numpy as np
import re
pd.set_option('Display.max_rows', 1000)
```

```python
df = pd.read_csv('zomato_restaurants_lisbon')
df.drop('ID', inplace=True, axis=1)
df
```

```python
df[df['cash_cards?'] == "How do we calculate cost for two?"]
```

```python
# Create a column with a numerical value for avg_cost
avg_cost = df['avg_cost']
avg_cost
df['avg_cost_2'] = [int(re.match('\d+', string).group()) for string in avg_cost]
```

```python
df
```

```python
for index, value in df['type'].iteritems():
    df.loc[index, 'type'] = value.replace("[", '').replace("]", '').replace("'", '')
for index, value in df['cuisine'].iteritems():
    df.loc[index, 'cuisine'] = value.replace("[", '').replace("]", '').replace("'", '')
for index, value in df['more_infos'].iteritems():
    df.loc[index, 'more_infos'] = value.replace("[", '').replace("]", '').replace("'", '')
```

```python
types = [value.split(', ') for value in df['type']]
df_types = df.loc[:, ['name', 'neighborhood', 'avg_rating', '#_of_ratings', 'avg_cost', 'avg_cost_2', 'cash_cards?']]
df_types['type1'] = [value[0] for value in types]
df_types['type2'] = [value[1] if len(value) > 1 else '' for value in types]               
```

```python
df_types = pd.melt(df_types, id_vars=['name', 'neighborhood', 'avg_rating', '#_of_ratings', 'avg_cost', 'avg_cost_2', 'cash_cards?'])
```

```python
df_types.drop(df_types[(df_types['value'] == '') & (df_types['variable'] == 'type2')].index, inplace=True)
df_types
```

```python
# Create a list with all the cuisines per restaurant and drop the last value for each sublist (erroneously imported the neighborhood name)
cuisines_list = [value.split(', ') for value in df['cuisine']]
for cuisines in cuisines_list:
    del cuisines[-1]
```

```python
df_cuisines = df.loc[:, ['name', 'neighborhood', 'avg_rating', '#_of_ratings', 'avg_cost', 'avg_cost_2', 'cash_cards?']]
for j in range(0, len(cuisines_list)):
    cuisines = cuisines_list[j]
    for i in range(0, len(cuisines)):
        df_cuisines.loc[j, 'cuisine' + str(i+1)] = cuisines[i]
df_cuisines
```

```python
# Melt the various cuisine columns
df_cuisines = pd.melt(df_cuisines, id_vars=['name', 'neighborhood', 'avg_rating', '#_of_ratings', 'avg_cost', 'avg_cost_2', 'cash_cards?'])
```

```python
# Drop the rows with Nan values
df_cuisines.drop(df_cuisines[(df_cuisines['value'].isnull()) & (df_cuisines['variable'] != 'cuisine1')].index, inplace=True)
df_cuisines
```
